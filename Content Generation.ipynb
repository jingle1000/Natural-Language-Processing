```python
from bs4 import BeautifulSoup
import itertools
```

```python
positive_reviews = BeautifulSoup(open('electronics/positive.review').read())
positive_reviews = positive_reviews.findAll("review_text")
#show an example review
positive_reviews[0].text.replace("\n", "")
```

```python
trigrams = {}

for review in positive_reviews:
    review = review.text.lower().replace("\n", "")
    review_tokens = review.split(" ")
    for i in range(len(review_tokens) - 2):
        key = (review_tokens[i], review_tokens[i+2])
        if key not in trigrams:
            trigrams[key] = []
        trigrams[key].append(review_tokens[i+1])
```

```python
trigram_probabilites = {}
for k, words in trigrams.items():
    if len(words) > 1:
        prob = {}
        n = 0
        for word in words:
            if word not in prob:
                prob[word] = 1
            else:
                prob[word] += 1
            n += 1
        for word, count in prob.items():
            prob[word] = float(count) / n
        trigram_probabilites[k] =  prob
        
```
